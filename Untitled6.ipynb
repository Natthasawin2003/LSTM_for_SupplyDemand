```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from google.colab import files
import matplotlib.dates as mdates
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('s1i10.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.set_index('date', inplace=True)
```

```python
df = df.resample('D').mean()
df
```

```python
plt.plot(df['sales'], label='Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Data Over Time')
plt.legend()
plt.show()
```

```python
df['sales_smoothed'] = df['sales'].rolling(window=365).mean()
plt.plot(df['sales_smoothed'], label='Smoothed Sales Data')
```

```python
df['sales'] = df['sales'].rolling(window=7, min_periods=1).mean()
plt.plot(df['sales'], label='Smoothed Sales Data')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['sales'], label='Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Data Over Time')
plt.legend()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format: Year-Month
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Set ticks at month intervals
plt.xticks(rotation=90)
plt.show()
```

```python
df['sales']
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df['sales'] = scaler.fit_transform(df[['sales']])
df['sales']
```

```python
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)
```

```python
sequence_length = 30
```

```python
train_size = int(len(df) * 0.8)
train_data = df['sales'].values[:train_size]
test_data = df['sales'].values[train_size - sequence_length:]
```

```python
train_data
```

```python
test_data
```

```python
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)
```

```python
X_train, y_train
```

```python
X_test, y_test
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
X_train
```

```python
X_test
```

```python

# Parameters to tune
batch_size = 16  # Try 16, 64 as well
epochs = 10  # Try increasing or decreasing based on results
lstm_units = 100  # Try 50, 150, or 200
dropout_rate = 0.2  # Adjust between 0.1 to 0.5
learning_rate = 0.001  # Try 0.0005 or 0.01 if needed

# Model Building
model = Sequential()
model.add(LSTM(units=lstm_units, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))
# model.add(Dropout(dropout_rate))
# model.add(LSTM(units=lstm_units, return_sequences=False))
model.add(Dropout(dropout_rate))
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test), verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
```

```python
train_predictions = scaler.inverse_transform(train_predictions)
y_train = scaler.inverse_transform([y_train])
test_predictions = scaler.inverse_transform(test_predictions)
y_test = scaler.inverse_transform([y_test])
```

```python
# Adjust train_data_len to account for sequence length
train_data_len = len(df) - len(test_predictions)

# Create a DataFrame to store predictions along with the actual values
predictions_df = df.copy()
predictions_df['Train'] = np.nan
predictions_df['Test'] = np.nan
predictions_df['Predicted'] = np.nan

# Fill the training data
predictions_df.iloc[:train_data_len, predictions_df.columns.get_loc('Train')] = scaler.inverse_transform(train_data.reshape(-1, 1)).flatten()

# Fill the actual test data and predicted data
predictions_df.iloc[train_data_len:, predictions_df.columns.get_loc('Test')] = scaler.inverse_transform(test_data[sequence_length:].reshape(-1, 1)).flatten()
predictions_df.iloc[train_data_len:, predictions_df.columns.get_loc('Predicted')] = test_predictions.flatten()

# Plotting
plt.figure(figsize=(14, 5))
plt.plot(predictions_df.index, predictions_df['Train'], label='Training Data', color='blue')
plt.plot(predictions_df.index, predictions_df['Test'], label='Actual Sales', color='green')
plt.plot(predictions_df.index, predictions_df['Predicted'], label='Predicted Sales', color='red')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('LSTM Predictions vs Actual Sales')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
```

```python
rmse = math.sqrt(mean_squared_error(y_test[0], test_predictions))
mae = mean_absolute_error(y_test[0], test_predictions)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
```
